```python
import json
import os
from pathlib import Path

target_folder = ['batch-1', 'batch-2']
```

```python

#parts   = ["part1", "part2", "part3", "part4", "part5", "part6", "part7"]


new_res = {}
for folder in target_folder:
    #for part in parts:
    src_file = f"../../REAL_DATA/Data/feature_metadata/{folder}/all_keyframes_mapping.json"
    print(src_file)
    
    if Path(src_file).exists() == False: continue
    with open(src_file, 'r') as f:
        data = json.load(f)

    for frame_data in data:
        image_path = frame_data['image_path']
        frame_id = frame_data['frame_id']
        video_id = frame_data['video_id']

        parent_dir = os.path.dirname(image_path)
        grandparent_dir = os.path.dirname(parent_dir)
        key_name = os.path.basename(grandparent_dir)

        # Use last 8 characters of video_id as key
        video_id_key = video_id[-8:]
        if key_name not in new_res:
            new_res[key_name] = {}

        if video_id_key not in new_res[key_name]:
            new_res[key_name][video_id_key] = []

        new_res[key_name][video_id_key].append(f"{frame_id}.webp")

with open("beit_grouped_keyframes_metadata.json", 'w') as f:
    json.dump(new_res, f)
```

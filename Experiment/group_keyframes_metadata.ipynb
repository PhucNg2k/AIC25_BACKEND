```python
import json
import os

taget_folder = ['vit-batch1', 'vit-batch2']
new_res = {}
for folder in taget_folder:
    src_file = f"../../REAL_DATA/{folder}/extracted_features/all_keyframes_mapping.json"

    with open(src_file, 'r') as f:
        data = json.load(f)

    
    for frame_data in data:
        image_path = frame_data['image_path']
        frame_id = frame_data['frame_id']
        video_id = frame_data['video_id']

        parent_dir = os.path.dirname(image_path)         # one level up
        grandparent_dir = os.path.dirname(parent_dir)    # two levels up
        key_name = os.path.basename(grandparent_dir)     # get the folder name
        if key_name not in new_res:
            new_res[key_name] = {}

        # Initialize list if video_id not present under key_name
        if video_id not in new_res[key_name]:
            new_res[key_name][video_id] = []

        # Append frame_id to the list
        new_res[key_name][video_id].append(f"{frame_id}.webp")
        

with open("grouped_keyframes_metadata.json", 'w') as f:
        json.dump(new_res, f)


    
```
